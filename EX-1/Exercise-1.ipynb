```python
import random as rnm
```

```python
k=rnm.randint(0,1)
```

```python
def Image_Random_Matrix(start,end):
    Image_Matrix=[]
    for i in range(0,8):
        Pixels=[]
        for j in range(0,8):
            k=rnm.randint(start,end)
            Pixels.append(k)
        Image_Matrix.append(Pixels)
    return Image_Matrix
```

```python
Image_Matrix_Black=Image_Random_Matrix(0,1)
```

```python
Image_Matrix_Black
```

```python
from PIL import Image
import numpy as np
```

```python
array = np.array(Image_Matrix_Black, dtype=np.uint8)
```

```python
array
```

```python
new_image = Image.fromarray(array)
new_image.save('new.png')
```

```python
Image_Matrix_GreyScale=Image_Random_Matrix(0,255)
```
