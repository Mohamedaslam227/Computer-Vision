```python
import cv2 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
image = cv2.imread('Aslam.jpg',0) 
```

```python

plt.subplot(1, 2, 1) 
plt.title("Original") 
plt.imshow(image) 
```

```python
alpha = 1.5 
beta = 50  
image2 = cv2.convertScaleAbs(image, alpha=alpha, beta=beta) 
```

```python
plt.subplot(1, 2, 2) 
plt.title("Brightness & contrast") 
plt.imshow(image2) 
plt.show() 
```

```python
kernel = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]]) 
sharpened_image = cv2.filter2D(image, -1, kernel) 
```

```python
plt.imshow(sharpened_image,cmap='gray') 
```

```python
_, simple_thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
```

```python
adaptive_thresh = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11,5)
```

```python
_, otsu_thresh = cv2.threshold(image, 0, 255,cv2.THRESH_OTSU)
```

```python
plt.imshow(simple_thresh,cmap='gray')
```

```python
plt.imshow(adaptive_thresh,cmap='gray')
```

```python
plt.imshow(otsu_thresh,cmap='gray')
```
