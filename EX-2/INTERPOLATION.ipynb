```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('Aslam.jpg')
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
img.shape
```

```python
LL=cv2.resize(img,(800,1100),interpolation=cv2.INTER_LINEAR)
```

```python
plt.imshow(cv2.cvtColor(LL,cv2.COLOR_BGR2RGB))
```

```python
LL=cv2.resize(img,(800,1100),interpolation=cv2.INTER_NEAREST)
```

```python
plt.imshow(cv2.cvtColor(LL,cv2.COLOR_BGR2RGB))
```

```python
LL=cv2.resize(img[1000:,600:],(2340*2, 1692*2),interpolation=cv2.INTER_CUBIC)
```

```python
plt.imshow(cv2.cvtColor(LL,cv2.COLOR_BGR2RGB))
```
