```python
import cv2
import random as rnm
import numpy as np
```

```python
def Image_Random_Matrix(start,end):
    Image_Matrix=[]
    for i in range(0,8):
        Pixels=[]
        for j in range(0,8):
              Pixels.append(i%2)
        Image_Matrix.append(Pixels)
    return Image_Matrix
```

```python
Image1=np.array(Image_Random_Matrix(0,1))
```

```python
Image2=np.array(Image_Random_Matrix(0,1))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(Image1,cmap='gray')
```

```python
plt.imshow(Image2,cmap='gray')
```

```python
AND=cv2.bitwise_and(Image1,Image2)
```

```python
OR=cv2.bitwise_or(Image1,Image2)
```

```python
XOR=cv2.bitwise_xor(Image1,Image2)
```
