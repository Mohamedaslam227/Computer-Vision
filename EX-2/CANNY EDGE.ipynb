```python
import cv2
import numpy as np
```

```python
img = cv2.imread('Aslam.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
import matplotlib.pyplot as plt
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,cmap='gray')
```

```python
blur=cv2.GaussianBlur(gray,(5,5),0.56)
```

```python
plt.imshow(blur,cmap='gray')
```

```python
edges=cv2.Canny(blur,threshold1=100,threshold2=200)
```

```python
plt.imshow(edges)
```

```python
edges=cv2.Canny(blur,threshold1=10,threshold2=150)
```

```python
plt.imshow(edges)
```

```python
edges=cv2.Canny(blur,threshold1=70,threshold2=110)
```

```python
plt.imshow(edges)
```
