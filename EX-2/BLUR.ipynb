```python
image=cv2.imread('Aslam.jpg')
```

```python
import cv2
import numpy as np
```

```python
Gb=cv2.GaussianBlur(image,(15,15),125)
```

```python
kernel = cv2.getGaussianKernel(15,125)
kernel_2d = np.outer(kernel, kernel.transpose())
```

```python
kernel_2d.shape
```

```python
mb=cv2.medianBlur(image,15)
```

```python
ab=cv2.blur(image,(15,15))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(Gb,cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(cv2.cvtColor(mb,cv2.COLOR_BGR2RGB))
```

```python
plt.imshow(kernel_2d, cmap='gray', interpolation='none')
plt.title('2D Gaussian Kernel')
plt.colorbar()
plt.show()
```

```python
plt.imshow(cv2.cvtColor(ab,cv2.COLOR_BGR2RGB))
```
